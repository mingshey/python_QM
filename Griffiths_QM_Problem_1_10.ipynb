```python
import numpy as np
import matplotlib.pyplot as plt
```

digits of $\pi$

$3.141592653589793238462643\dots$



```python
pi_digits25 = np.array([3,1,4,1,5,9,2,6,5,3,5,8,9,7,9,3,2,3,8,4,6,2,6,4,3])
```

```python
len(pi_digits25)
```

```python
d = np.arange(0,10)
print(d)
```

```python
np.linspace(d[0]-.5, d[-1]+.5, len(d)+1)
```

$N(d), (0\le d\le9)$

```python
Nd, bins = np.histogram(pi_digits25, bins=np.linspace(d[0]-.5, d[-1]+.5, len(d)+1))
print(Nd)
```

$$N=\sum_{d=0}^{9} N(d)$$

```python
N = Nd.sum()
print(f"N = {N}")
```

```python
plt.grid()
plt.bar(d, Nd)
```

$P(d) = \dfrac{N(d)}{N}$

```python
Pd = Nd/len(pi_digits25)
print(Pd)
```

$$\left<d\right> = \sum_{d=0}^{9} d \cdot P(d)$$

```python
d_avg = (d * Pd).sum()
print(f"<d> = {d_avg:.2f}")
```

$$\left<d^2\right> = \sum_{d=0}^{9} d^2 P(d)$$

```python
dsq_avg = (d**2 * Pd).sum()
print(r"<d²> = "+f"{dsq_avg:.2f}")
```

$\sigma_d^2 = \left<d^2\right> - \left<d\right>^2$

```python
d_var = dsq_avg - d_avg**2
print(f"σ² = {d_var:.2f}")
```

$\sigma_d = \sqrt{\left<d^2\right> - \left<d\right>^2}$

```python
d_sigma = np.sqrt(d_var)
print(f"σ = {d_sigma:.2f}")
```
