# Griffiths QM, 3rd ed.
## Problem 1.13 numerical experiment

Probability density of a simple harmonic oscillator
with potential defined as $V(x) = \dfrac{1}{2} k x^2$ and with energy $E$.

The amplitude of oscillation $b={\sqrt{\dfrac{2E}{k}}}$ is treated as $1$ here.

```python
import numpy as np
import matplotlib.pyplot as plt
```

* Numerical experiment

```python
def x(t):
  return np.cos(t)
```

```python
snapshot = x(np.random.rand(10000) * np.pi)
```

```python
plt.grid()
plt.ylim(0,2)
plt.hist(snapshot, bins=30, density=True);
```

* Analytical solution

Problability density: ($b=\sqrt{2E/k}=1$)
$$\rho(x) = \frac{1}{\pi \sqrt{b^2 - x^2}}$$

```python
def rho(x):
  return 1 / (np.pi * np.sqrt(1 - x**2))
```

```python
xs = np.linspace(-1, 1, 1000)
ys = rho(xs)
plt.grid()
plt.ylim(0,2)
plt.plot(xs, ys)
```

```python
plt.grid()
plt.ylim(0, 2)
plt.xlabel('x')
plt.ylabel(r'$\rho(x)$')
plt.title(r'Probability density $\rho(x)$')

plt.hist(snapshot, bins=30, density=True, label='Numerical experiment', color='green', alpha=0.5)
plt.plot(xs, ys, label='Analytical solution', c='k')

plt.legend()
```
