# Griffiths QM
## 2.3.2 Analytic method

```python
import numpy as np
from scipy.special import hermite
import matplotlib.pyplot as plt
```

```python
import math
```

```python
from scipy.integrate import quad, simpson
```

```python
import sympy as sym
```

```python
n = 10
```

```python
H_n = hermite(n, monic=False)
```

```python
H_n
```

```python
b = np.sqrt(2*n+1)
```

```python
x = np.linspace(-2*b, 2*b, 1000)
y = H_n(x) * np.exp(-(x**2)/2) / np.sqrt(np.sqrt(np.pi) * math.factorial(n)) / np.sqrt(2**n)
plt.plot(x, y)
```

```python
ysqr = y**2
plt.plot(x, ysqr)
```

```python
prob = lambda x: 1 / (np.pi * np.sqrt((1 - x**2/(2*n+1))) * np.sqrt(2*n+1) )
```

```python
plt.ylim([0, 1.2 * ysqr.max()])
plt.grid()
plt.plot(x, ysqr)
plt.plot(x, prob(x))
```

```python
print(f'{quad(prob, -b, b)[0]:f}')
```

```python
print(f'{simpson(ysqr, x=x):f}')
```
