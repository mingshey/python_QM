* Number of Samples

```python
N = Nj.sum()
print(N)
```

* Probability

$P(j) = \dfrac{N(j)}{N}$

```python
Pj = Nj / N
Pj
```

* Histogram of Probability

```python
plt.grid()
plt.bar(j, Pj, width=1-1/20)
```

* Check if Normalized

```python
sumPj = Pj.sum()
print(sumPj)
```

* average

$j N(j)$

```python
jNj = j * Nj
jNj
```

$$\left< j \right> = \dfrac{1}{N}\sum_{j=0}^{\infty} j N(j)$$

```python
javg = jNj.sum() / N
print(javg)
```

$j P(j)$

```python
jPj = j * Pj
jPj
```

$$\left< j \right> = \sum_{j=0}^{\infty} j P(j)$$

```python
javg = jPj.sum()
print(javg)
```

## Problem 1.2

(a)

$j^2$

```python
jsq = j ** 2
print(jsq)
```

$$\left<j^2\right>=\sum_{j=0}^{\infty} j^2 P(j)$$

```python
jsq_avg = (j**2 * Pj).sum()
print(jsq_avg)
```

$\left<j\right>^2$

```python
javg_sq = javg**2
print(javg_sq)
```

(b)

$\Delta j = j - \left< j \right>$

```python
Delta_j = j - javg
print(Delta_j)
```

$\left<\Delta j ^2\right> = \sum \Delta j ^2 P(j)$

```python
Delta_j_sqr = Delta_j**2
print(Delta_j_sqr)
```

```python
vari = (Delta_j_sqr * Pj).sum()
print(vari)
```

```python
stdev = np.sqrt(vari)
print(stdev)
```

(c)

* Variance

$\sigma^2 = \left<j^2\right> - \left<j\right>^2$

```python
sigma_sq = jsq_avg - javg_sq
print(sigma_sq)
```

* Standard Deviation

$\sigma = \sqrt{\left<j^2\right> - \left<j\right>^2}$

```python
sigma = np.sqrt(sigma_sq)
print(sigma)
```
