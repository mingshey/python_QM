```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, c=1):
  return np.exp(-x) - c * x + 1

def df(x, c=1):
  return -np.exp(-x) - c

```

```python
def nr(f, df, x0=1, c=1, tol=1e-6, max_iter=100):
  '''
  Newton-Raphson method for finding roots of f(x, c) = 0
  Inputs:
    f: function f(x, c)
    df: derivative of f(x, c) with respect to x
    x0: initial guess
    c: parameter
  '''
  x = x0
  for i in range(max_iter):
    x_new = x - f(x, c) / df(x, c)
    if abs(x_new - x) < tol:
      return x_new
    x = x_new
  return x
```

```python
nr(f, df, x0=1, c=1)
```

```python
z1 = nr(f, df, x0=1, c=1/2)
z1
```

```python
z2 = nr(f, df, x0=1, c=2)
z2
```

```python
def g(x, c=1):
  return np.exp(-x) - 1 + c*x

def dg(x, c=1):
  return -np.exp(-x) + c
```

```python
nr(g, dg, x0=1, c=1)
```

```python
nr(g, dg, x0=1, c=2)
```

```python
z3 = nr(g, dg, x0=1, c=0.5)
z3
```

```python
def f1(x):
  return np.exp(-x)
```

```python
def g1(x, c=1):
  return c * x - 1

def g2(x, c=1):
  return 1 - c * x
```

```python
xs = np.linspace(0, 5, 1000)
y1 = f1(xs)
y2 = g1(xs, c=1/2)
y3 = g1(xs, c=2)
y4 = g2(xs, c=1/2)
y5 = g2(xs, c=1)
y6 = g2(xs, c=2)
```

```python
plt.ylim(-2,2)
plt.grid()
plt.plot(xs, y1, label=r'$e^{-x}$')
plt.plot(xs, y2, label=r'$x/2 - 1$')
plt.plot(xs, y3, label=r'$2 x - 1$')
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.scatter(z1, f1(z1), color='red', label='z1')
plt.scatter(z2, f1(z2), color='green', label='z2')
plt.xlabel('z')
plt.legend()
plt.show()
```

```python
plt.ylim(-2,2)
plt.grid()
plt.plot(xs, y1, label=r'$e^{-x}$')
plt.plot(xs, y4, label=r'$1 - x/2$')
plt.plot(xs, y5, label=r'$1 - x$', ls=':')
plt.plot(xs, y6, label=r'$1 - 2x$')
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.scatter(z3, f1(z3), color='red', label='z3')
plt.xlabel('z')
plt.legend()
plt.show()
```
