# Griffiths, _Introduction to Quantum Mechanics_
Problem 1.4

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 1
a = 1
b = 2

def Psi(x, A, a, b):
  if 0 <= x <= a:
    return A * (x/a)
  elif a <= x <= b:
    return A * (b-x)/(b-a)
  else:
    return 0

#Psi = np.vectorize(Psi)
```

```python
x = np.linspace(-0.5, 4.5, 1000)

y = np.array([Psi(i, A, a, b) for i in x])
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('$\Psi(x)$')
plt.title('Wavefunction $\Psi(x)$')
plt.show()
```

```python
psi_sq = y ** 2
plt.plot(x, psi_sq)
plt.xlabel('x')
plt.ylabel('$\Psi^2(x)$')
plt.title('Probability density $\Psi^2(x)$')
plt.show()
```

```python
Asqr = 1 / np.trapz(psi_sq, x)
print(f'A^2 = {Asqr}')
A1 = np.sqrt(Asqr)
print(f'A = {A1}')
```

```python
y1 = np.array([Psi(i, A1, a, b) for i in x])
plt.plot(x, y1)
```

```python
psi1_sq = y1 ** 2
plt.plot(x, psi1_sq)
```

```python
P2 = np.sqrt(np.trapz(psi1_sq, x))
print(f'Ptot = {P2:.3f}')
```
